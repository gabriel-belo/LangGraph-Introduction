```python
from typing import TypedDict

class StateGraph(TypedDict):
    gaph_state: str
```

```python
def node1(state: StateGraph)-> StateGraph:
    print("--------Node 1--------")
    return  {"graph_state": state["graph_state"] +  "I am"}

def node2(state:StateGraph)->StateGraph:
    print("--------Node 2--------")
    return {"graph_state": state["state_graph"] + " happy!"}

def node3(state: StateGraph)-> StateGraph:
    print("--------Node 2--------")
    return {"graph_state": state["state_graph"] + " sad!"}
```

```python
import random
from typing import Literal

def decide_mood(state)-> Literal["node2", "node3"]:

    user_input= state["graph_state"]

    if random.random() < 0.5:
        return "node2"

    return "node3"
```

```python
from langgraph.graph import StateGraph, START, END

builder= StateGraph(StateGraph)

builder.add_node("node1", node1)
builder.add_node("node2", node2)
builder.add_node("node3", node3)


builder.add_edge(START, "node1")
builder.add_conditional_edges("node1", decide_mood)
builder.add_edge("node2", END)
builder.add_edge("node3", END)


graph= builder.compile()
```
